```python
#23.7
!pip install osmnx folium

import osmnx as ox
import networkx as nx
import folium
from IPython.display import display

ox.settings.use_cache = True
ox.settings.log_console = False

place = "District 1, Ho Chi Minh City, Vietnam"

G = ox.graph_from_place(place, network_type="drive")

diem_A = (10.7769, 106.6953)
diem_B = (10.7700, 106.7000)

node_A = ox.distance.nearest_nodes(G, diem_A[1], diem_A[0])
node_B = ox.distance.nearest_nodes(G, diem_B[1], diem_B[0])

route_dijkstra = nx.shortest_path(G, node_A, node_B, weight="length")

def heuristic(u, v):
    y1, x1 = G.nodes[u]["y"], G.nodes[u]["x"]
    y2, x2 = G.nodes[v]["y"], G.nodes[v]["x"]
    return ((y1 - y2)**2 + (x1 - x2)**2)**0.5

route_astar = nx.astar_path(G, node_A, node_B, heuristic=heuristic, weight="length")

def get_length(graph, route):
    tong = 0
    for u, v in zip(route[:-1], route[1:]):
        data = graph.get_edge_data(u, v)
        if data is not None:
            canh_ngan_nhat = min(data.values(), key=lambda x: x["length"])
            tong += canh_ngan_nhat["length"]
    return tong

len_dijkstra = get_length(G, route_dijkstra)
len_astar = get_length(G, route_astar)

print("Chiều dài Dijkstra (m):", round(len_dijkstra, 2))
print("Chiều dài A* (m):", round(len_astar, 2))

m = folium.Map(location=diem_A, zoom_start=14)

coords_dijkstra = [(G.nodes[n]["y"], G.nodes[n]["x"]) for n in route_dijkstra]
coords_astar = [(G.nodes[n]["y"], G.nodes[n]["x"]) for n in route_astar]

folium.PolyLine(
    coords_dijkstra,
    color="blue",
    weight=5,
    tooltip="Dijkstra"
).add_to(m)

folium.PolyLine(
    coords_astar,
    color="red",
    weight=3,
    tooltip="A*"
).add_to(m)

folium.Marker(
    diem_A,
    popup="Điểm A",
    icon=folium.Icon(color="green")
).add_to(m)

folium.Marker(
    diem_B,
    popup="Điểm B",
    icon=folium.Icon(color="red")
).add_to(m)

display(m)

print()
print("NHẬN XÉT")
print("- Dijkstra tìm đường đi ngắn nhất chính xác nhưng thường duyệt nhiều nút hơn.")
print("- A* dùng heuristic nên thường tìm nhanh hơn.")
print("- Nếu heuristic phù hợp thì kết quả A* và Dijkstra thường rất gần nhau hoặc giống nhau.")
print("- Trong giao thông thông minh, A* phù hợp cho tìm đường thời gian thực.")
```

```python
#23.12
!pip install folium geopy

import folium
import random
from geopy.distance import geodesic
from IPython.display import display

kho = [
    {"ten": "Kho 1", "toa_do": (10.7769, 106.6953)},
    {"ten": "Kho 2", "toa_do": (10.7820, 106.7000)}
]

don_hang = [
    (10.7800,106.6980),
    (10.7790,106.6940),
    (10.7750,106.6900),
    (10.7720,106.6920),
    (10.7710,106.6970),
    (10.7830,106.7030)
]

def tim_kho_gan_nhat(diem):
    return min(kho, key=lambda k: geodesic(diem, k["toa_do"]).km)

def tao_tuyen_toi_uu(start, points):
    chua_di = points.copy()
    hien_tai = start
    tuyen = [start]
    tong = 0

    while chua_di:
        diem_gan = min(chua_di, key=lambda p: geodesic(hien_tai, p).km)
        tong += geodesic(hien_tai, diem_gan).km
        hien_tai = diem_gan
        tuyen.append(diem_gan)
        chua_di.remove(diem_gan)

    return tuyen, tong

def tao_tuyen_ngau_nhien(start, points):
    ds = points.copy()
    random.shuffle(ds)
    tuyen = [start] + ds
    tong = 0

    for i in range(len(tuyen)-1):
        tong += geodesic(tuyen[i], tuyen[i+1]).km

    return tuyen, tong

ban_do = folium.Map(location=[10.7769,106.6953], zoom_start=14)

tong_opt = 0
tong_random = 0

for k in kho:
    diem_kho = k["toa_do"]
    ds_khach = [d for d in don_hang if tim_kho_gan_nhat(d) == k]

    if not ds_khach:
        continue

    tuyen_opt, kc_opt = tao_tuyen_toi_uu(diem_kho, ds_khach)
    tuyen_rand, kc_rand = tao_tuyen_ngau_nhien(diem_kho, ds_khach)

    tong_opt += kc_opt
    tong_random += kc_rand

    folium.PolyLine(tuyen_opt, color="green", weight=5, tooltip="Tối ưu").add_to(ban_do)
    folium.PolyLine(tuyen_rand, color="red", weight=3, tooltip="Ngẫu nhiên").add_to(ban_do)

    folium.Marker(diem_kho, popup=k["ten"], icon=folium.Icon(color="blue")).add_to(ban_do)

    for d in ds_khach:
        folium.CircleMarker(d, radius=5, color="black", fill=True).add_to(ban_do)

display(ban_do)

print("Tổng quãng đường (tối ưu):", round(tong_opt,2), "km")
print("Tổng quãng đường (ngẫu nhiên):", round(tong_random,2), "km")

print()
print("NHẬN XÉT")
print("- Tuyến tối ưu (màu xanh) sử dụng heuristic chọn điểm gần nhất nên giảm quãng đường di chuyển.")
print("- Tuyến ngẫu nhiên (màu đỏ) thường dài hơn do không có chiến lược lựa chọn.")
print("- Việc tối ưu tuyến giúp giảm chi phí vận chuyển và thời gian giao hàng.")
print("- Đây là ứng dụng của AI/heuristic trong bài toán logistics thực tế.")
```

```python
#23.15
!pip install folium scikit-learn

import folium
import numpy as np
from sklearn.cluster import KMeans
from IPython.display import display

khach_hang = np.array([
    [10.7765,106.6950],
    [10.7770,106.6960],
    [10.7780,106.6970],
    [10.7800,106.7000],
    [10.7810,106.7010],
    [10.7820,106.7020],

    [10.7720,106.6900],
    [10.7710,106.6890],
    [10.7700,106.6880],

    [10.7850,106.7050],
    [10.7860,106.7060],
    [10.7870,106.7070]
])

k = 3
model = KMeans(n_clusters=k, random_state=0)
labels = model.fit_predict(khach_hang)
centers = model.cluster_centers_

ban_do = folium.Map(location=[10.7769,106.6953], zoom_start=14)

colors = ["red","blue","green"]

for i in range(len(khach_hang)):
    folium.CircleMarker(
        location=khach_hang[i],
        radius=5,
        color=colors[labels[i]],
        fill=True,
        fill_opacity=0.7,
        popup=f"Khách - cụm {labels[i]}"
    ).add_to(ban_do)

for i, c in enumerate(centers):
    folium.Marker(
        location=c,
        popup=f"Kho đề xuất {i}",
        tooltip="Kho tối ưu",
        icon=folium.Icon(color="black", icon="home")
    ).add_to(ban_do)

display(ban_do)

print("Vị trí kho đề xuất:")
for i, c in enumerate(centers):
    print(f"Kho {i}: ({c[0]:.4f}, {c[1]:.4f})")

print()
print("NHẬN XÉT")
print("- Dữ liệu khách hàng được phân cụm bằng K-Means.")
print("- Tâm cụm là vị trí tối ưu để đặt kho trung chuyển.")
print("- Giúp giảm khoảng cách giao hàng và chi phí vận hành.")
print("- Ứng dụng thực tế trong logistics, gọi xe, giao hàng nhanh.")
```
